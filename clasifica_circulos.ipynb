```python
# Standard imports
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Setup SimpleBlobDetector parameters.
params = cv2.SimpleBlobDetector_Params()

# Change thresholds
params.minThreshold = 10;
params.maxThreshold = 200;

# Filter by Area.
params.filterByArea = True
params.minArea = 1500

# Filter by Circularity
params.filterByCircularity = True
params.minCircularity = 0.1

# Filter by Convexity
params.filterByConvexity = True
params.minConvexity = 0.87

# Filter by Inertia
params.filterByInertia = True
params.minInertiaRatio = 0.01

# Create a detector with the parameters
detector = cv2.SimpleBlobDetector_create(params)

# Read image
img = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circulo4503684241162240.jpg', cv2.IMREAD_GRAYSCALE)

# Detect blobs.
keypoints = detector.detect(img)

# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the     circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(img, keypoints, np.array([]),         (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show keypoints
cv2_imshow(im_with_keypoints)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

import os
from os import listdir

# img1 = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circulo4503684241162240.jpg', 0)
# img2 = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circulo4503611113472000.jpg', 0)
# img3 = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circulo4503690750722048.jpg', 0)
# img4 = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circulo4503747759702016.jpg', 0)
img_cp = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/circPerfecta.jpg', 0) 

# get the path/directory
folder_dir = '/content/drive/My Drive/datos/circulos_prueba'
for image_N in os.listdir(folder_dir):
 
  # check if the image ends with jpg
  if (image_N.endswith(".jpg")):
        print(image_N)

        image = cv2.imread(f'/content/drive/My Drive/datos/circulos_prueba/'+image_N, 0)
        #res = cv2.matchTemplate(image,img_cp,cv2.TM_CCOEFF_NORMED)
        #w, h = img_cp.shape[::-1]
        #print("res:"+ str(res))

        ret, thresh = cv2.threshold(image, 127, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        #obtener contornos
        contours, hierarchy = cv2.findContours(thresh, cv2.RETR_CCOMP , cv2.CHAIN_APPROX_NONE)
        #dibujo contornos
        cv2.drawContours(image, contours, -1, (155, 111, 10), 1)
        cv2_imshow(image)
        cv2.waitKey(0)

        #obtener perimetro(arclength)
        arclength = cv2.arcLength(contours[0],True);
        approx = cv2.approxPolyDP(contours[0],0.01*arclength,True)
        print ("approxPolyDP:"+str(len(approx)))

        #calculo area a partir de los contornos
        if len(contours) > 0:
            area1 = cv2.contourArea(contours[0])
        else: #no contours --> perfect circle, black image, no image,...
            print ("Sorry No contour Found.")


        #obtener circulo exterior
        (x,y),radius = cv2.minEnclosingCircle(contours[0])
        center = (int(x),int(y))
        radius = int(radius) 

        
        circularity_v2 = 4 * np.pi * area1 / (arclength * arclength);
 

        circularity_v1 = 1 - (abs(area1-(np.pi*radius*radius))/(np.pi*radius*radius))

        difAreas=(np.pi*radius*radius)-area1

        print("circularity_v1:"+ str(circularity_v1))
        print("circularity_v2:"+ str(circularity_v2))
        print("difAreas:"+ str(difAreas))
        print("arclength:"+ str(arclength))
```
